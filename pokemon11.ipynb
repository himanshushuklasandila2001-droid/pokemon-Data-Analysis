```python
import matplotlib.pyplot as plt
import numpy as py
import pandas as pd
```

```python
df = pd.read_csv("pokemon.csv")
df
```

```python
plt.plot(df["Generation"], df["Speed"].sort_values(),
color="red",
marker="x",
markerfacecolor="red",
markersize=9,
label="Generation",
alpha=0.8)


plt.show()
```
